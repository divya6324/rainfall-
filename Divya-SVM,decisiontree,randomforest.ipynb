```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=["PrecipitationSumInches"])
y=df["PrecipitationSumInches"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## SVM

```python
from sklearn.svm import SVR
```

```python
model=SVR()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred[0:6]
```

```python
df1=pd.DataFrame(y_pred,y_test)
df1
```

```python
from sklearn.metrics import mean_squared_error
```

```python
testing_error=mean_squared_error(y_test,y_pred)
testing_error
```

```python
rmse=np.sqrt(testing_error)
rmse
```

```python
#Residual sum of squares
rss=np.sum(np.square(y_test-y_pred))
rss
```

```python
from sklearn.metrics import r2_score
```

```python
r2_error_testing=r2_score(y_test,y_pred)
r2_error_testing
```
